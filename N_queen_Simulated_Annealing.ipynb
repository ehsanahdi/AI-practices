# Defining the cost function 

```python
import random # for each queen position
# we want it to return a full list of random queens
import math
import time
def random_queen(N):
    return [random.randint(0,N-1) for _ in range(N)]

# in this def we want to see how many of queens in each situation attack each other 
def conflict(state):
    conflicts = 0
    N = len(state)
    for i in range(N):
        for j in range(i+1, N):
            
            if state[i] == state[j] or abs(state[i] - state[j]) == abs(i-j):
                conflicts += 1
    return conflicts
# using two dimensional for loop let us to g trough each cell in N * m board

```

# Geneerating neighbor states
in each col we change the queen position , hoping for finding a better place !


```python
def neighbors(state):
    N = len(state)
    new_state = state.copy() # copy for not changing the original
    col = random.randint(0,N-1) # the queen we move (random)
    new_row = random.randint(0,N-1) # the random new row position
    new_state[col] = new_row # moving queen to new random place
    return new_state
```

# Main part of simulated Annealing 

defining the core logic and the main def

set the temp to 0.1 so decrase so fast !

when delta is negative -> the board is better

```python
def simulated_annealing(
        N,
        initial_temp = 100,
        cooling_rate=0.99,
        min_temp = 0.1,
        reheating = 1.2,
        wait = 100
    ):
    # initialize with random config 
    current_state = random_queen(N)
    current_energy = conflict(current_state)
    best_state = current_state
    best_energy = current_energy

    T = initial_temp
    itr = 0
    start_time = time.time()
    is_energy_better = 0 # for keeping track of improvements -> Reheating
    energy = [] # for log of energy values over iterations
    

    # till we have good temp or not find the answer we continue
    while T > min_temp and current_energy > 0 :
        itr += 1
        neighbor = neighbors(current_state)
        neighbor_energy = conflict(neighbor)
        delta = neighbor_energy - current_energy
        #calculate the Delta for compairing the neighbor with current one 
        
        # if we wnat to accept the neighbor (less than0 delta is good): 
        if delta < 0 or random.random()  < math.exp(- delta / T) :
            current_state = neighbor
            current_energy = neighbor_energy
            
        # this we have is good so far 
        if current_energy < best_energy :
            best_state = current_state
            best_energy = current_energy
            is_energy_better = 0 # reset the counter
        else:
            is_energy_better += 1
            energy.append(current_energy)
            
            if is_energy_better >= wait:
                T *= reheating
                is_energy_better = 0
                print(f"Reheating at iteration {itr}, Temperature: {T:.2f}")
            
        # it's time for coooling down
        T *= cooling_rate
    end_time = time.time()

    runtime = end_time - start_time
        
        
    return best_state , best_energy , itr,runtime
```

# Def for printing the board 

using . for other board and "Q" for queens

```python
def print_board(state):
    N = len(state)
    for i in range(N):
        row = ""
        for j in range(N):
            row += "Q " if state[j] == i else ". "
        print(row)
    print()
    
```

```python
if __name__ == "__main__" : 
    N = 8
    best_state, best_energy, itr, runtime = simulated_annealing(N)
    print(" best confing fouond in", itr, "itration with", best_energy , "conflicts:")
    print("Runtime :", runtime , "seconds")
    print_board(best_state)
```
