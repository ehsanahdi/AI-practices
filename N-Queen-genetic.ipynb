# Genetic algorithm for N-queen probelem

defining the random chromosome for starting point and 
fitness for findinf the conflicts count 
our purpse is to have the most fitness as possible , so
we have reduce the conflicts

```python
import random  
import math
import time

def random_chromosome(N):
    return [random.randint(0, N-1) for _ in range(N)]

def fitness(chromosome):
    N = len(chromosome)
    conflicts = 0
    total_pairs = N * (N - 1) / 2
    for i in range(N):
        for j in range(i + 1, N):
            if chromosome[i] == chromosome[j] or abs(chromosome[i] - chromosome[j]) == abs(i - j):
                conflicts += 1
    non_atacking = total_pairs - conflicts
    return non_atacking
    
    
```

# Creating parents for next generations
We choose 3 of them (K=3)
Crossover , choosing two of them for creating next gen

```python
def selection(population, fitnessess , k=3): # 3 is the number of parents which chooseby accident from population
    selecteds = random.sample(list(zip(population, fitnessess)), k)
    selecteds.sort(key= lambda x :x[1], reverse=True)
    return selecteds[0][0]


def crossover(parent1, parent2):
    N = len(parent1)
    point = random.randint(1, N - 2)
    child = parent1[:point] + parent2[point:]
    return child
```

# For not stocking in local optimization
We have to change it randomly so it not stock

```python
def mutate(chromosome, mutation_rate=0.01):
    N = len(chromosome)
    new_chromosome = chromosome.copy()
    if random.random() < mutation_rate:
        col = random.randint(0, N - 1)
        new_chromosome[col] = random.randint(0, N - 1)
    return new_chromosome

def count_conflicts(chromosome):
    N = len(chromosome)
    conflicts = 0
    for i in range(N):
        for j in range(i+1, N):
            if chromosome[i] == chromosome[j] or abs(chromosome[i] - chromosome[j]) == abs(i-j):
                conflicts += 1
    return conflicts
    
```

# Final part


```python
def genetic(N, population_size=100, generations=1000, mutation_rate=0.01,tournament_size=3):
    start = time.time()
    
    population = [random_chromosome(N) for _ in range(population_size)] # creating first generation
    best_chromosome = None
    best_fitness = -1
    fitness_progress = []
    
    for gen in range(1, generations+1):
        fitnesses = [fitness(chrom) for chrom in population]
        
        max_fitness = max(fitnesses)
        best = fitnesses.index(max_fitness)
        if max_fitness > best_fitness:
            best_fitness = max_fitness
            best_chromosome = population[best]
                    
        fitness_progress.append(best_fitness)
        
        total_pairs = N * (N - 1) / 2
        if best_fitness == total_pairs:
            print(f"Solution found in generation {gen}")
            break
        
        new_population = []
        for _ in range(population_size):
            parent1 = selection(population, fitnesses)
            parent2 = selection(population, fitnesses)
            child = crossover(parent1, parent2)
            child = mutate(child, mutation_rate)
            new_population.append(child)
        population = new_population
        
        
    end = time.time()
    run_time = end - start
    best_conflicts = count_conflicts(best_chromosome)  
    return best_chromosome, best_fitness, best_conflicts, gen, run_time

```

```python
def print_board(state):
    N = len(state)
    for i in range(N):
        row = ""
        for j in range(N):
            row += "Q " if state[j] == i else ". "
        print(row)
    print()
```
