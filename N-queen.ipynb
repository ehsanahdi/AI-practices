# def "Is safe" check that if i put the queen , is this safe or not , relative to other queens
N -> board chess size
board -> holding current selection

```python
def is_queen_safe(row, col):
    global board, N  # because of Jupyter and I want to be accessible and explain each part separately
    # The reason we iterate this is that we need to check previous columns (0 -> col-1)
    for back_col in range(col):
        back_row = board[back_col] 
        # start from first column and go to col-1 
        
        if back_row == row:          
            return False

        if abs(back_row - row) == abs(back_col - col):  
            return False 

    return True  
```

 solving n_queens using backtracking


```python
def place(col):
    global board, N, solutions
    if col == N:  # for stopping recursion â€” means all queens placed well
        solutions.append(board.copy())
        return
    
    for row in range(N):  # each time try to place the queen in one row 
        if is_queen_safe(row, col):
            board[col] = row
            place(col + 1)
            board[col] = -1  # backtrack

```

in here we check that the placed queen is safe from rows,col , and is there any cnoflict or not ?!

```python
def solve_queen(n):
    global solutions, board, N
    N = n
    solutions = []
    board = [-1] * N
        # board[i] = row position of queen in column i
        
    place(0)
    return solutions
        
```

```python

N = 4
solutions_a = solve_queen(N)

print(f"Found {len(solutions_a)} solutions for N = {N}:\n")
for sol in solutions_a:
    print(sol)
```
